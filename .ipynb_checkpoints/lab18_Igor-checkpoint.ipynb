```python
df_source = pd.read_csv('/Users/igorhufnagel/Desktop/Ironhack/LABS/Week 8/Day 2/lab-cleaning-categorical-data/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
column_to_extract = df_source['Effective To Date']
df_destination = pd.read_csv('/Users/igorhufnagel/Desktop/Ironhack/LABS/Week 9/Day 2/lab-feature-extraction/categorical.csv')
df_destination['effective_to_date'] = column_to_extract
df_destination.to_csv('/Users/igorhufnagel/Desktop/Ironhack/LABS/Week 9/Day 2/lab-feature-extraction/categorical.csv', index=False)
```

```python
df_destination.head()
```

```python
categoricals_new = df_destination
```

```python
categoricals_new['effective_to_date'] = pd.to_datetime(categoricals_new['effective_to_date'], format='%m/%d/%y')
```

```python
import re

year_pattern = r"\d{4}"

month_pattern = r"-\d{2}-" 

categoricals_new['year_effective_to_date'] = categoricals_new['effective_to_date'].map(lambda x: re.findall(year_pattern,str(x))[0] 
                if len(re.findall(year_pattern,str(x))) > 0 else np.nan)
categoricals_new['month_effective_to_date'] = categoricals_new['effective_to_date'].map(lambda x: re.findall(month_pattern,str(x))[0][1:-1] 
                if len(re.findall(month_pattern,str(x))) > 0 else np.nan)
```

```python
categoricals_new
```

```python
categoricals_new.set_index('effective_to_date')

plt.plot(categoricals_new.index, categoricals_new['effective_to_date'])
plt.xlabel('Value')
plt.ylabel('Effective to Date')
plt.title('Time Series')
plt.show()
```
