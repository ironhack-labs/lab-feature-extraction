```python
# Importing pandas library
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing data
df_customer = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Labs\lab-cleaning-categorical-data\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")

# Standardizing columns names
cols = []
for i in range(len(df_customer.columns)):
    cols.append(df_customer.columns[i].lower().replace(" ","_"))
df_customer.columns = cols
df_customer

# Creating a new dataframe with categorical values
df_categ = df_customer.select_dtypes(include = object)
df_categ = df_categ.drop(["customer"], axis = 1)
df_categ.head()
```

```python
# 2) Plot all the categorical variables with the proper plot. What can you see?

# Each plot shows the absolute frequency of the attribute of each category

for col in df_categ.columns:
    if col != "effective_to_date":
        sns.histplot(df_categ[col])
        plt.xticks(rotation = 20)
        plt.show()
```

```python
# 3) There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?

# Policy_type column is redundant because policy column gives us also that information. So we can dropp it.

print(df_categ["policy_type"].unique())
print(df_categ["policy"].unique())

df_categ = df_categ.drop(["policy_type"], axis = 1)
df_categ
```

```python
# 4) Plot time variable.
plot = sns.countplot(df_categ, x = "effective_to_date")
plt.xticks(rotation = 90)
plt.show()

# Can you extract something from it?

# Not really, that's why I decided to group the effective_to_date by the respective month in [lab-cleaning-categorical-data]:

df_categ["effective_to_date"] = df_categ["effective_to_date"].str.split("/",expand=True)[0]
df_categ.loc[df_categ["effective_to_date"] == "1", "effective_to_date"] = "jan/2011"
df_categ.loc[df_categ["effective_to_date"] == "2", "effective_to_date"] = "fev/2011"
df_categ.rename(columns = {"effective_to_date":"effective_to_date_(month)"}, inplace = True)

sns.histplot(df_categ["effective_to_date_(month)"])
plt.show()
```
