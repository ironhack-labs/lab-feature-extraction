# 2. Plot all the categorical variables with the proper plot. What can you see?

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as pamella
```

```python
data.columns
```

```python
# Define the columns to plot
cols_to_plot = ['Customer', 'State', 'Response', 'Coverage', 'Education', 'EmploymentStatus', 'Gender', 'Location Code', 'Marital Status', 'Policy Type', 'Policy', 'Renew Offer Type', 'Sales Channel', 'Vehicle Class', 'Vehicle Size']

# Create the figure and axes
fig, axs = pamella.subplots(5, 3, figsize=(25, 30))

# Loop through the columns to plot
for i, col in enumerate(cols_to_plot):

    # Select the appropriate axes
    ax = axs[i//3, i%3]

    # Create a countplot
    sns.countplot(x=col, data=data, ax=ax)

    # Set the axis label and tick label font size
    ax.set_xlabel(col, fontsize=14)
    ax.tick_params(labelsize=12)
    
    # Rotate x-axis tick labels to vertical and y-axis tick labels to horizontal
    ax.tick_params(axis='x', rotation=90)
    ax.tick_params(axis='y', rotation=0)

# Add a title to the figure
fig.suptitle('Distribution of Categorical Variables', fontsize=20)

# Adjust the spacing between subplots
fig.tight_layout(pad=5)

# Show the plot
pamella.show()
```

```python
<p>the customer plot doesn't show nothing<\p>
<p>we can see that the most costumer are from california and oregon<\p>
<p>we can see that the most costumer didn't respond<\p>
<p>we can see that the most costumer are employed<\p>
<p>we can see that the most costumer preffer medsize and foor door vehicles<\p>
```
