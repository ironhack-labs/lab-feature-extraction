```python
#1. Open the categoricals variable we created before.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

categorical=pd.read_csv("categorical.csv",index_col=0)
categorical
```

```python
#2. Plot all the categorical variables with the proper plot. What can you see?

for column in categorical.columns:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

    categorical[column].value_counts().plot.bar(ax=ax1)
    ax1.set_title(column)

    categorical[column].value_counts().plot.pie(ax=ax2, autopct="%1.1f%%")
    ax2.set_title(column)

    fig.tight_layout()
    plt.show()
```

It is visible that in some columns, We could join some values into "Other" be more comparable to others, and decrease the number of variables. E.g., in policy joining all the "Special", employment status be "employed", "unemployed" and "others", and so on. In this case let's leave those as they are, as they might be important.

```python
categorical["effective_to_date"] = pd.to_datetime(categorical["effective_to_date"])
categorical["effective_year"] = categorical["effective_to_date"].dt.year
categorical["effective_month"] = categorical["effective_to_date"].dt.month
categorical["effective_month"] = categorical["effective_to_date"].dt.strftime("%B") # Converts month number to month name

categorical
```

```python
categorical.drop(columns=["effective_to_date"],axis=1,inplace=True)
```
