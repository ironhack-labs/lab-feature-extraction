```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('categorical.csv', index_col=0)
data.head()
```

```python
#Plot all the categorical variables with the proper plot. What can you see?
# I see no pattern, like a normal distribution.

for column in data.columns:
    data[column].value_counts().plot(kind='bar')
    plt.title(column)
    plt.show()

```

```python
#There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?
# Yes, vehicle size just brings almost same data as vehicle type, and renew offer brings same data as sales channel.
# I will drop them.

data = data.drop('vehicle_size', axis=1)
```

```python
data = data.drop('renew_offer_type', axis=1)
```
